```python
# 1. solicitamos la edad del usuario
edad = int(input("Porfavor introduce tu edad:"))
# 2. Evaluamos si cumple con los rangos de tarifa preferencial

if edad < 0:
    print("Edad inválida. Por favor, introduce un número positivo.")
elif edad < 12:
    print("Tienes una tarifa infantil (preferencial).")
elif edad >= 65:
    print("Tienes una tarifa de la tercera edad (preferencial).")
else:
    print("Tienes una tarifa estándar.")
```
